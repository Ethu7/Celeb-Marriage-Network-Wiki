```python
import pandas as pd
import networkx as nx

%matplotlib inline

import seaborn as sns
sns.set_style('darkgrid')
```

```python
marriages = pd.read_csv('testing.csv')
marriages = marriages[[x for x in marriages.columns if 'Spouse' in x]]

marriages.columns = ['source', 'target']

G = nx.from_pandas_edgelist(marriages)
```

```python
positions = nx.spring_layout(G)
```

```python


df = pd.DataFrame(positions).T

sns.scatterplot(0, 1, data = df)
```

```python
eigen = nx.eigenvector_centrality(G)
between = nx.betweenness_centrality(G)
close = nx.closeness_centrality(G)
```

```python
eigendf = pd.Series(eigen, name = 'Eigen')
betweendf = pd.Series(between, name = 'Between')
closedf = pd.Series(close, name = 'Closeness')

centralitydf = pd.concat([eigendf, betweendf, closedf], axis = 1)
centralitydf.head()
```

```python
ax = sns.scatterplot('Eigen', 'Closeness', data = centralitydf)
```

```python
centralitydf['Eigen'].sort_values(ascending = False)
```
